```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 1.1 - Load the `./data/college_completion.csv` data with Pandas.
df = pd.read_csv("./data/college_completion.csv")
```

```python
# 1.2 - What are are the dimensions of the data? How many observations are there? What are the variables included? 
# Use `.head()` to examine the first few rows of data.

print("Dimensions:", df.shape[0], "x", df.shape[1], "\n")
print("Observations:", df.shape[0], "\nVariables:", df.shape[1], "\n")
print("First 15 Variables:")
for col in df.columns[:15]:
    print(col)

df.head()
```

```python
# 1.3 - Cross tabulate `control` and `level`. Describe the patterns you see.
pd.crosstab(df["control"], df["level"])

# I see that there are slightly more 4-year private for-profit institutions than 2-year ones
# I also see that of not-for-profit private schools, 4-year makes up the vast majority
# As for public, there are more 2-year than 4-year schools
```

```python
# 1.4 - For `grad_100_value`, create a histogram, kernel density plot, boxplot, and statistical description.
gv = df["grad_100_value"]

plt.figure()
plt.hist(gv, color = 'navy')
plt.xlabel("Grad 100 Value")
plt.ylabel("Frequency")
plt.show()

plt.figure()
sns.kdeplot(gv, color = "navy")
plt.xlabel("Grad 100 Value")
plt.ylabel("Density")
plt.show()

plt.figure()
sns.boxplot(gv, color = "navy")
plt.ylabel("Grad 100 Value")
plt.show()

gv.describe()
```

```python
# 1.5 - For `grad_100_value`, create a grouped kernel density plot by `control` and by `level`. 
# Describe what you see. Use `groupby` and `.describe` to make grouped calculations of statistical descriptions of `grad_100_value` by `level` and `control`. 
# Which institutions appear to have the best graduation rates?

fig, axs = plt.subplots(1, 2, figsize = (10, 6), sharey = True)
axs = axs.flatten()
vars = ['control', 'level']
for ax, var in zip(axs, vars):
    sns.kdeplot(data = df, x = "grad_100_value", hue = var, palette = "magma", ax = ax)
    sns.despine(ax = ax, top = True, right = True)
    ax.tick_params(axis = "y", length = 0)
    ax.grid(True, axis = 'y', alpha = 0.75, color = 'lightgray')

plt.show()
```

```python
df.groupby("level")["grad_100_value"].describe()
```

```python
df.groupby("control")["grad_100_value"].describe()
```

In the KDE by control plot, I notice an extremely high public density at low grad_100_values, with private not-for-profit the highest at greater grad_100_values, followed by private for-profit, then public. In the KDE by level plot, 4-year schools have slightly better results, with a higher density than 2-year school at higher grad_100_values. Looking at the description grouped by level, 4-year's mean is higher by about 12, and the median is higher by about 17, supporting the conclusion from the plot about 4-year schools having better graduation rates. When grouped by control, private not-for-profit schools lead the way with a mean around 42, followed by private for-profit schools at 29, and public schools at 17. Overall, private not-for-profit and 4-year schools have the best graduation rates.

```python
# 1.6 - Create a new variable, `df['levelXcontrol']=df['level']+', '+df['control']` that interacts level and control.
# Make a grouped kernel density plot. Which institutions appear to have the best graduation rates?
df['levelXcontrol'] = df['level'] + ', ' + df['control']
plt.figure()
sns.kdeplot(data = df, x = "grad_100_value", hue = "levelXcontrol", palette = "magma")
plt.show()

# 4-year, Private not-for-profit institutions appear to have the best graduation rates, followed by 2-year, Private not-for-profit, and 2-year, Private for-profit institutions.
```

```python
# 1.7 - Make a kernel density plot of `aid_value`. Now group your graph by `level` and `control`. 
# What explains the shape of the graph? 
# Use `groupby` and `.describe` to make grouped calculations of statistical descriptions of `aid_value` by `level` and `control`.

plt.figure()
sns.kdeplot(data = df, x = 'aid_value', color = "navy")
plt.show()

fig, axs = plt.subplots(1, 2, figsize = (10, 6), sharey = True)
axs = axs.flatten()
vars = ['control', 'level']
for ax, var in zip(axs, vars):
    sns.kdeplot(data = df, x = "aid_value", hue = var, palette = "magma", ax = ax)
    sns.despine(ax = ax, top = True, right = True)
    ax.tick_params(axis = "y", length = 0)
    ax.grid(True, axis = 'y', alpha = 0.75, color = 'lightgray')

plt.show()
```

The shape of the graph is caused by the second bumps in the Private not-for-profit and 4-year institutions, which contribute to the shape of the total KDE around 10,000-20,000

```python
df.groupby("level")["aid_value"].describe()
```

```python
df.groupby("control")["aid_value"].describe()
```

```python
# 1.8 - Make a scatterplot of `grad_100_value` by `aid_value`. Describe what you see. 
# Now make the same plot, grouping by `level` and then `control`. 
# Describe what you see. For which kinds of institutions does aid seem to increase graduation rates?

plt.figure()
sns.scatterplot(data = df, x = 'aid_value', y = 'grad_100_value', color = "navy", s = 12)
plt.xlabel("Aid Value")
plt.ylabel("Grad 100 Value")
plt.show()

# In the overall plot, I see two groups, one which goes vertically at low aid values, and one that increases like a square root.

plt.figure()
sns.scatterplot(data = df, x = 'aid_value', y = 'grad_100_value', hue = 'level', palette = 'magma', s = 12)
plt.xlabel("Aid Value")
plt.ylabel("Grad 100 Value")
plt.show()

plt.figure()
sns.scatterplot(data = df, x = 'aid_value', y = 'grad_100_value', hue = 'control', palette = 'magma', s = 12)
plt.xlabel("Aid Value")
plt.ylabel("Grad 100 Value")
plt.show()
```

Aid value seems to have a significant impact on 4-year and private not-for profit institutions. For the other types, the scatter plots move up in a fairly straight line, with aid having little effect on graduation rates, but the points for 4-year and private not-for-profit institutions are low graduation rates at low aid values and higher graduation rates at high aid values.
